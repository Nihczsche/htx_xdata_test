## HTX xData Test cv-hotword-5a Python Notebook

This notebook provides the solutions for detecting hot words from the transcribed results from cv-valid-dev mp3 dataset using the finetuned model in task 4. 

The hot words to be detected are: "be careful", "destroy" and "stranger". The list of mp3 filenames with the hot words detected are saved into detected.txt.

```python
import pandas as pd
import os

# Load the cv-valid-dev.csv file generated from Q4 into a dataframe
cv_dev_metadata = pd.read_csv(os.path.join('..', 'asr-train', 'cv-valid-dev.csv'))

cv_dev_metadata.head(5)
```

```python
# Find strings which contain the phrases given in the hotword list
hotword_list = ["DESTROY", "BE CAREFUL", "STRANGER"]

cv_dev_exactmatch_generated = cv_dev_metadata[cv_dev_metadata['finetuned_text'].str.contains('|'.join(hotword_list), na=False)]

print(len(cv_dev_exactmatch_generated))
cv_dev_exactmatch_generated
```

```python
# Save the mp3 filenames with hotwords detected into detected.txt file
import numpy as np

np.savetxt('detected.txt', cv_dev_exactmatch_generated['filename'].values, fmt='%s')
```
